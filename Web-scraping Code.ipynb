```python
import requests
from requests import get
from bs4 import BeautifulSoup
from warnings import warn
import numpy as np
import pandas as pd
import html5lib
import lxml
```

```python
user_agent = "Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.37"
base_url = 'https://www.the-numbers.com/'

url = 'https://www.the-numbers.com/market/2022/top-grossing-movies'
data = requests.get(url, headers={'User-Agent': user_agent})
html_code = BeautifulSoup(data.text, 'html.parser')

df = pd.DataFrame()
```

```python
titles = []
dates = []
company = []
genres = []
gross = []
tickets = []
months = []
days = []
urls = []
ratings_temp = []
ratings = []
times_temp = []
```

```python
#MOVIE URLS
movie_tags = html_code.find_all('b')
urls = []
for tag in movie_tags:

    ini_str = str(tag)
    sub_str = "\""
    occurrence = 1

    start = -1
    for i in range(0, occurrence):
      start = ini_str.find(sub_str, start + 1)

    sub_str = "\""
    occurrence = 2

    end = -1
    for i in range(0, occurrence):
      end = ini_str.find(sub_str, end + 1)

    ini_str = ini_str[start : end]
    sub_str = ">"
    occurrence = 3

    carrot = -1
    for i in range(0, occurrence):
      carrot = ini_str.find(sub_str, carrot + 1)

    new = ini_str[carrot + 2 : end]
    new = base_url + new
    #print(new)
    urls.append(new)

urls.pop(382)
urls.pop(381)
urls.pop(380)
urls.pop(379)
urls.pop(378)
urls.pop(0)
```

```python
#MOVIE TITLE
titles = []

for tag in movie_tags:
    if tag.find('a') is not None:
        title = tag.a.text
        titles.append(str(title))
        
    else:
        titles.append(None)
        
titles.pop(382)
titles.pop(381)
titles.pop(380)
titles.pop(379)
titles.pop(378)
#titles.pop(377)
#titles.pop(376)
titles.pop(0)

df["Movie Titles"] = titles
```

```python
#RATINGS STEP 1
i = 0
ratings_temp = []
for link in urls:
    
    url = link
    data = requests.get(url, headers={'User-Agent': user_agent})
    html_code = BeautifulSoup(data.text, 'html.parser')
    
    rating_tags = html_code.find_all('a')
    sub_str = "market/mpaa-rating/"
   
    if sub_str in str(html_code):
        i = i + 1
        #print(i)
        for tag in rating_tags:
            ini_str = str(tag)
            sub_str = "market/mpaa-rating/"
            if sub_str in ini_str:
                #print(ini_str)
                ratings_temp.append(ini_str)
                break;
                
    else:
        i = i + 1
        #print(i)
        none = "N/A"
        #print(none)
        ratings_temp.append(none)
            
```

```python
#FINAL RATINGS
ratings = []
for tag in ratings_temp:

    ini_str = str(tag)
    sub_str = "market/mpaa-rating/"

    if sub_str in ini_str:

        sub_str = "</a>"
        occurrence = 1

        end = -1
        for i in range(0, occurrence):
            end = ini_str.find(sub_str, end + 1)
            #print(end)


        sub_str = ">"
        occurrence = 1
            
        middle = -1
        for i in range(0, occurrence):
            middle = ini_str.find(sub_str, middle + 1)
            #print(middle)


        new = ini_str[middle + 1 : end]
        #print(new)
        ratings.append(new)
            
    else:
        none = "N/A"
        #print(none)
        ratings.append(none)
        
df["MPAA Rating"] = ratings
```

```python
#RUNNING TIME STEP 1
i = 0
times_temp = []
for link in urls:

    url = link
    data = requests.get(url, headers={'User-Agent': user_agent})
    html_code = BeautifulSoup(data.text, 'html.parser')
    
    rating_tags = html_code.find_all('td')
    sub_str = "minutes"
   
    if sub_str in str(html_code):
        i = i + 1
        print(i)
        
        for tag in rating_tags:
            ini_str = str(tag)
            sub_str = "minutes"
            
            if sub_str in ini_str:
                #print(ini_str)
                times_temp.append(ini_str)
                break;
       
    else:
        i = i + 1
        #print(i)
        none = "N/A"
        #print(none)
        times_temp.append(none)

```

```python
#FINAL RUNNING TIME
times = []
for tag in times_temp:
    if tag != "N/A":
        times.append(int(tag[4:7]))
    else:
        times.append(tag)
        
times.insert(372, "N/A" )

df["Running Time (Min)"] = times
```

```python
#RELEASE DATE STEP 1
url = 'https://www.the-numbers.com/market/2022/top-grossing-movies'
data = requests.get(url, headers={'User-Agent': user_agent})
html_code = BeautifulSoup(data.text, 'html.parser')

date_tags = html_code.find_all('a')

dates_temp = []
i = 0
sub_str = "/daily/"

for tag in date_tags:

    if sub_str in str(tag):
        #print(tag)
        dates_temp.append(tag)
```

```python
#RELEASE DATE - FINAL
dates = []
for tag in dates_temp:
    ini_str = str(tag)
    sub_str = ">"
    occurrence = 1

    start = -1
    for i in range(0, occurrence):
      start = ini_str.find(sub_str, start + 1)

    sub_str = "</a>"
    occurrence = 1

    end = -1
    for i in range(0, occurrence):
      end = ini_str.find(sub_str, end + 1)

    ini_str = ini_str[start : end]
    sub_str = "<"
    occurrence = 2

    carrot = -1
    for i in range(0, occurrence):
      carrot = ini_str.find(sub_str, carrot + 1)

    new = ini_str[carrot+2 : end]

    dates.append(new)

df["Release Date"] = dates
```

```python
#GENRE 
genres = []
x = 0
#sub_str = "/market/distributor/"
for tag in comp_tags:
    sub_str = "/market/genre/"
    
    if sub_str in str(tag):
        
        
        ini_str = str(tag)
        sub_str = "/market/genre/"
        occurrence = 1
        start = -1
        for i in range(0, occurrence):
          start = ini_str.find(sub_str, start + 1)

        
        sub_str = "</a>"
        occurrence = 4
        end = -1
        for i in range(0, occurrence):
          end = ini_str.find(sub_str, end + 1)
        

        ini_str = ini_str[start : end]
        sub_str = ">"
        occurrence = 1
        carrot = -1
        for i in range(0, occurrence):
          carrot = ini_str.find(sub_str, carrot + 1)

        
        new = ini_str[carrot + 1 : end]
        
        #x = x + 1
        #print(x)
        #print(new)
        genres.append(new)
        
        
    else:
        x = x + 1
        #print(x)
        #print("N/A")
        genres.append("N/A")

#genres[147] = "Comedy"
genres[148] = "Action"
genres[196] = "Thriller/Suspense"
genres[239] = "Comedy"
genres[326] = "Action"
genres[373] = "Multiple Genres"

df["Genre"] = genres
```

```python
#GROSS MONEY FINAL
gross = []
x = 0
for tag in comp_tags:
    ini_str = str(tag)
    sub_str = ">$"
    occurrence = 1

    start = -1
    for i in range(0, occurrence):
      start = ini_str.find(sub_str, start + 1)

    #Printing nth occurrence
    #print ("Nth occurrence is at", start)


    # FIND 2nd OCCURENCE OF </a>
    sub_str = "</td>"
    occurrence = 6

    end = -1
    for i in range(0, occurrence):
      end = ini_str.find(sub_str, end + 1)

    #Printing nth occurrence
    #print ("Nth occurrence is at", end)


    #FIND 1st OCCURENCE OF >
    ini_str = ini_str[start : end]
    sub_str = ">"
    occurrence = 1

    carrot = -1
    for i in range(0, occurrence):
      carrot = ini_str.find(sub_str, carrot + 1)

    #Printing nth occurrence
    #print ("Nth occurrence is at", carrot)


    new = ini_str[carrot+1 : end]
    new = new.replace('$', '')
    new = new.replace(',', '')
    new = new.replace(' ', '')
    #x = x + 1
    #print(x , ':')
    gross.append((new))

gross[376] = '347'
df["2022 Gross"] = gross
```

```python
#TICKETS FINAL
tickets = []
for tag in comp_tags:
    ini_str = str(tag)
    sub_str = "\"data\">"
    occurrence = 3

    start = -1
    for i in range(0, occurrence):
      start = ini_str.find(sub_str, start + 1)


    sub_str = "</td>"
    occurrence = 7

    end = -1
    for i in range(0, occurrence):
      end = ini_str.find(sub_str, end + 1)


    ini_str = ini_str[start : end]
    sub_str = ">"
    occurrence = 1

    carrot = -1
    for i in range(0, occurrence):
      carrot = ini_str.find(sub_str, carrot + 1)


    new = ini_str[carrot+1 : end]
    new = new.replace('$', '')
    new = new.replace(',', '')
    new = new.replace(' ', '')
    #print(new)
   
    tickets.append(new)

tickets[376] = '347'
df["Tickets Sold"] = tickets
```

```python
#make month and day columns
# initializing string
months = []
days = []
for i in dates:
    month = i[0:3]
    month = month.replace(' ', '')

    day = i[4:6]
    day = day.replace(' ', '')
    day = day.replace(',', '')
    
    months.append(month)
    days.append(day)

df['Release Month'] = months
df['Release Day'] = days

```
